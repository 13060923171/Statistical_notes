```python
import numpy as np
```

```python
# 随机生成两个样本
x = np.random.randint(0, 100, 100)
y = np.random.randint(0, 100, 100)
```

```python
print(x)
```

```python
print(y)
```

```python
# 计算平均值
x_mean = x.mean()
y_mean = y.mean()
```

```python
print(x_mean,y_mean)
```

```python
#求方差
x_var = np.var(x)
```

```python
print(x_var)
```

```python
#求y方差
y_var = np.var(y)
print(y_var)
```

```python
# 计算标准差
stdx = x.std()
stdy = y.std()
```

```python
# 计算协方差矩阵
covxy = np.cov(x, y)
print(covxy)
```
